```python
import matplotlib.pyplot as plt
import csv

f = open('seoul2.csv')
data = csv.reader(f)
next(data)
result = []
result2 = []
for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '01' and row[0].split('-')[2] == '23' :
            result.append(float(row[-1]))
            result2.append(float(row[-2]))

plt.rc('font')
plt.title('생일날의 기온')
plt.plot(result)
plt.plot(result2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul2.csv')
data = csv.reader(f)
next(data)
result = []
result2 = []
for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '01' :
            result.append(float(row[-1]))
        if row[0].split('-')[1] == '08' :
            result2.append(float(row[-1]))
            
plt.rc('font')
plt.title('서울 지역 1월, 8월 최고 기온 데이터')
plt.hist(result, bins = 100, label = 'JAN')
plt.hist(result2, bins = 100, label = 'AUG')
plt.legend()
plt.show()
```
